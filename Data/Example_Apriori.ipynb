```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
```

```python
df = pd.read_csv('Example_Apriori.csv')
```

```python
df
```

```python
for i in range(len(df)):
    df['Items'][i] = df['Items'][i].split(', ')
df
```

```python
df['Items'].str.join('|')
```

```python
oht = df['Items'].str.join('|').str.get_dummies()
oht
```

```python
frequent_itemsets = apriori(oht, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
association_rules_df = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
association_rules_df
```
